```python
# Prevent special characters like & and < to cause the browser to display something other than what you intended.
def html_escape(text):
    return html.escape(text)
```

```python
statement = """We believe the best companies are those that not only serve their customers, but reinvest in the communities where we all live and work.  We care for the uninsured by investing in local nonprofit clinics. We provide financial support to the state's health endowment fund in order to improve health and protect the most vulnerable of our neighbors. And we invest in improving thehealth of our children through grants to local schools. In the current crisis, we provide our customers with free coronavirus testing, treatment and vaccinations. We believe our customers are best served by serving the communities we call home."""
statement = statement.replace("\n", " ").split(" ")
statement
```

```python
df = pd.read_csv("mic2wav39.csv")
```

```python
filter_col = [col for col in df if str(col).startswith('HIGHLTE4')]
df = df[filter_col]
full_counts = pd.Series(df.values.ravel('F'))
```

```python
count_dictionary = full_counts.value_counts().to_dict()
max(list(count_dictionary.keys()))
```

```python
sums = []
for i in range(1, len(statement) + 1):
    sums.append(count_dictionary.get(i, 0))
```

```python
normalized_weights = [(float(i)-min(sums))/(max(sums)-min(sums)) for i in sums]
```
